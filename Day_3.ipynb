There are several types of modules in Python:
    
- Built-in modules

These are modules that are included with Python by default, such as the math, os, and sys modules.


- Third- party modules

These are modules created by other developers that can be installed using package managers such as pip. Examples include numpy, pandas, and matplotlib.


- User-defined modules

These are modules that you create yourself, typically stored in a file with a .py extension. These modules can be 

- sub-packages

These are modules that are organized in a package hierarchy, meaning a package can contain other sub-packages and modules. This allows for a better organization and management of the codebase.

```python
import numpy.random as random

# generate random integers
print(random.randint(1,100))
```

- Dynamic module

These are modules that are not stored in a file and are generated on the fly. These are typically used for plugin systems where a module is loaded only when it is needed.


```python
import importlib

# import module dynamically
module_name = "math"
math = importlib.import_module(module_name)

# use the sin function from the math module
print(math.sin(math.pi / 2))
```

```python
# Built-in module in  Python

import math 

#use the pi constant from the  math module

print(math.pi)

#use the sin() function from the math module 

result = math.sin(math.pi/2)
print(result)
```

- In Python, a library is a collection of modules that work together to provide additional functionality to a program, while a module is a single file containing Python definitions and statements. A library contains multiple modules that provide specific functionality. It can be imported as a whole, and then individual modules can be used as needed. A module, on the other hand, is a single file containing Python definitions and statements, it can contain classes, functions, and variables, and it can be imported and used in another Python program. A library is a group of modules that are related to each other and are used to achieve a specific functionality..
